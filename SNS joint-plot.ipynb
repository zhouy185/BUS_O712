## The jointplot

The jointplot extends the scatterplot by showing the distribution of the variables (x and y) along the axes.


* The basic command is given as `sns.jointplot(data=data_frame,x=column_a,y=column_b)`
* By default, it shows histograms to visualize the distributions

* The keyword argument `marginal_kws=dict(bins=n)` will use $n$ bins for the histogram


* We may use the argment `kind=` to change the type of the plot
    * For example, `kind='reg'` will make a scatterplot with a trend line


* We may again use `hue=` to color points by a categorical variable

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg',marginal_kws=dict(bins=30))

plt.suptitle("Tip vs. Total Bill")
plt.xlabel("Total bill (dollars)")
plt.ylabel("Tip (dollars)")
plt.show()
```
