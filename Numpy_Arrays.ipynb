## Arrays of evenly spaced values

We often want to use an evenly-spaced array (e.g., in creating graphs discussed later).

If we want to create an array starting with 0 and ending with 9, we can use `np.arange(start_value,end_value)`. 

Note that the first point is included and the last point isn't!

If we start from the value 0, it can be omitted.

`linspace(start_value, end_value)` also creates an evenly-spaced array.

Compared with `arange()`:
* The stop point is included in the range!
* We can specify how many values will be in the range (by default, 50 values)

```python
even_array = np.linspace(10,20)
even_array 
```

```python
even_array = np.linspace(10,20,10)
even_array 
```
