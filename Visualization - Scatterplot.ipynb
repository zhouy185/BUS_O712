The `scatter()` function creates scatterplots. Like a line graph, a scatterplot requires two variables, one to be placed on the x-axis and the other one on the y-axis. The basic codes are shown as follows.

`import matplotlib.pyplot as plt`

`plt.scatter(xAxis, yAxis)`

The data represents the population as well as murder rate (per 100,000 residents) and high school graduation rate (percentage) for each of 50 US states. Note that in the column "Region", each number in the set $\{1,2,3,4\}$ represents one of the following regions:

* 1: Northeast
* 2: Midwest
* 3: West
* 4: South

**Question**: Is murder rate correlated with the high school graduation rate?

To answer the question, let us create a scatterplot between 'Grad.Rate' (on the horizontal axis) and 'Murder.rate' (vertical axis).

```python
import matplotlib.pyplot as plt

plt.scatter(crime_df['Grad.Rate'],crime_df['Murder.Rate'],c=crime_df['Region'])
plt.title('Murder rate vs. high school graduation rate')
plt.xlabel('High school graduation rate')
plt.ylabel('Murder rate')
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(data=crime_df,x='Grad.Rate',y='Murder.Rate',hue='Region',size='Population',
                palette='bright',style='Region')
plt.title('Murder rate vs. high school graduation rate')
plt.xlabel('High school graduation rate')
plt.ylabel('Murder rate')
plt.show()
```

```python
NE_states = [state.strip() 
             for state in "Connecticut, Maine, Massachusetts, New Hampshire, Rhode Island, Vermont,New Jersey, New York, Pennsylvania".split(",")]


MW_states = [state.strip()
             for state in "Illinois, Indiana, Michigan, Ohio, Wisconsin,Iowa, Kansas, Minnesota, Missouri, Nebraska, North Dakota, South Dakota".split(",")]

S_states = [state.strip()
    for state in "Delaware, Florida, Georgia, Maryland, North Carolina, South Carolina, Virginia, \
            Washington, D.C., West Virginia, Alabama, Kentucky, Mississippi, Tennessee, Arkansas, Louisiana, Oklahoma, Texas".split(",")]

W_states = [state.strip()
           for state in "Arizona, Colorado, Idaho, Montana, Nevada, New Mexico, Utah, Wyoming, Alaska, California, Hawaii, Oregon, Washington".split(",")]

all_states = NE_states+MW_states+S_states+W_states
state_reg = ['Northeast']*len(NE_states)+['Midwest']*len(MW_states)+['South']*len(S_states)+['West']*len(W_states)

dict_state_reg = dict(zip(all_states,state_reg))
```

```python
dict_state_reg
```

```python
crime_df['Region']=[dict_state_reg[st] for st in crime_df['State']]
```
