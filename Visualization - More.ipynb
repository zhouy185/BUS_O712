## Plotting styles

The library matplotlib has multiple styles we can choose from. By default, matplotlib uses the style 'default'. However, you can choose your plotting style from nearly 30 different styles. See an array containing style names by running plt.style.available (note that the style 'default' is not included, as these are all the alternative styles that you can use). 

Each style has a different unique default colours, line style and colours, background format and colours, etc. You can review different matplotlib styles at [Style sheets reference â€” Matplotlib 3.5.3 documentation](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html).

To choose any of the other styles, simply run 
* `plt.style.use('name-of-style')`

This will use the selected style when creating plots in your active session. Your style resets to 'default' when you start a new file (session).

Sometimes we may want to include multiple plots in one figure.



The `plt.subplot()` function combines several plots into one figure, and has three arguments, `nrows`, `ncols`, and `index`. 

The argument of `plt.subplot(nrows, ncols, index)` includes integers, which determines the layout of the subplots.
* `nrows` indicates the number of rows. 
* `ncols` indicates the number columns  
* `index` indicates which spot is being used by the current plot.
* The comma between the three input arguments are often ommitted
