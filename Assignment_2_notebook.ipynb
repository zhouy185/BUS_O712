```python
import plotly.express as px

df_natadataset_copy = df_natadataset_copy.dropna(subset=['Income', 'NumWebVisitsMonth', 'NumWebPurchases'])
df_natadataset_copy['AcceptedCmp'] = (
    df_natadataset_copy[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']]
    .max(axis=1))

fig = px.scatter(
    df_natadataset_copy,
    x='NumWebVisitsMonth',
    y='NumWebPurchases',
    size='Income',
    color='AcceptedCmp',
    hover_data=['Income', 'Education', 'Marital_Status'],
    title='Frequency of visits to the website vs purchases online',
    size_max=30)

fig.show()
```

**Hint**: You can use the `size_max=` argument to adjust the scale of the point sizes and make the size differences easier to see.

```python
df_natadataset_copy['AcceptedCmpOverall'] = (
    df_natadataset_copy[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']]
        .max(axis=1))

df_long = df_natadataset_copy.melt(
    id_vars=['AcceptedCmpOverall'],
    value_vars=['NumWebVisitsMonth', 'NumWebPurchases'],
    var_name='ActivityType',
    value_name='Count')

fig = px.box(
    df_long,
    x='ActivityType',
    y='Count',
    color='AcceptedCmpOverall',
    title='Frequency of visits to the website vs purchases online',
    labels={
        'ActivityType': 'Customer Behaviour',
        'Count': 'Count per Month',
        'AcceptedCmpOverall': 'Accepted Campaign (1=Yes, 0=No)'})

fig.show()
```

```python
df_natadataset_plot = df_natadataset_copy.dropna(subset=[
    'Income', 'NumWebPurchases', 'MntWines', 'MntFruits',
    'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']).copy()

df_natadataset_plot['Total_Spending'] = (
df_natadataset_plot['MntWines'] +
df_natadataset_plot['MntFruits'] +
df_natadataset_plot['MntMeatProducts'] +
df_natadataset_plot['MntFishProducts'] +
df_natadataset_plot['MntSweetProducts'] +
df_natadataset_plot['MntGoldProds'])

df_natadataset_plot['AcceptedCmpOverall'] = (
df_natadataset_plot[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']]
    .max(axis=1))

fig = px.scatter(
    df_natadataset_plot,
    x='Income',
    y='Total_Spending',
    size='NumWebPurchases',
    color='AcceptedCmpOverall',
    hover_data=['Income', 'Education', 'Marital_Status'],
    title='Wealthier Customers vs Overall Spending',
    size_max=20,
    opacity=0.7,
    color_discrete_map={0: 'purple', 1: 'green'},
    range_x=[0, 120000])

fig.update_layout(
    xaxis_title="Income",
    yaxis_title="Total Spending Across All Products",
    legend_title="Accepted Campaign (1 = Yes, 0 = No)",
    plot_bgcolor='white')

fig.show()
```

### Task B.1 (15 points)

i. Consider that customers are grouped by their **education levels**. Develop a bar chart with several sets of bars, in which:

* Each set of bars (i.e.,bar of the same color) represent the spendings by customers in different groups on a spefic product category. For example, spending by customers "Graduation", "Master", "PhD", ..., on *fish products*.  
* The chart should allow you to easily visuallize both the **total spending** by customers in a group on all product categories and the **composition** of their spending on those product categories.

ii. Further, answer the following questions in a **text (markdown) cell** and make sure to explain your answer based on the chart:

**Hint**: By default, each row in the dataset is shown as a separate rectangle in the bar, and these rectangles have borders that can make the chart look cluttered. In that case, use `fig.update_traces(opacity=1, marker_line_width=0)` to remove the border lines.

```python
df_barchart_natadataset = df_natadataset_copy.copy()

spending_columns = [
    'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

df_spend = df_barchart_natadataset[['Education'] + spending_columns]

df_long = df_spend.melt(
    id_vars='Education',
    value_vars=spending_columns,
    var_name='Product_Category',
    value_name='Spending')

fig = px.bar(
    df_long,
    x='Education',
    y='Spending',
    color='Product_Category',
    title='Customer Spending vs Education Level',
    barmode='stack')

fig.update_traces(opacity=1, marker_line_width=0)
fig.update_layout(
    xaxis_title="Education Level",
    yaxis_title="Total Spending",
    legend_title="Product Category",
    plot_bgcolor='white')

fig.show()
```

* Within each group, the chart should allow you to easily compare the spending of customers on different product categories (e.g., idenfity the highest spending category by single customers).

ii. Further, answer the following questions in a **text (markdown) cell** and make sure to explain your answer based on the chart:
* For each marital status group, which category of product do they spend the highest amount on?
* Does the marital status affect customers' spending distribution across the product categories?  How?

```python
df_barchart_natadataset_marital = df_natadataset_copy.copy()

spending_columns = [
    'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

df_spend = df_barchart_natadataset_marital[['Marital_Status'] + spending_columns]

df_long = df_spend.melt(
    id_vars='Marital_Status',
    value_vars=spending_columns,
    var_name='Product_Category',
    value_name='Spending')

total_spending = (
    df_long.groupby('Marital_Status')['Spending']
    .sum()
    .sort_values(ascending=False)
    .index)

fig = px.bar(
    df_long,
    x='Marital_Status',
    y='Spending',
    color='Product_Category',
    title='Customer Spending vs Marital Status (Sorted by Total Spending)',
    barmode='group',
    category_orders={'Marital_Status': total_spending})

fig.update_traces(opacity=1, marker_line_width=0)

fig.update_layout(
    xaxis_title="Marital Status",
    yaxis_title="Total Spending",
    legend_title="Product Category",
    plot_bgcolor='white')

fig.show()
```

In this part, let us investigate variability in customers' online purchasing behavior.

i. We will use a visualization chart that shows distribution of customers' online purchases. The chart should:
* Show multiple distributions, each being the distribution of number of online purchases by customers of a certain generation (Gen X vs. Millennials,  ignore other generations)
* Allow users to observe the 1st, 2nd, and 3rd quartiles of the distribution, along with any outliers.

ii. Based on the chart, answer the following question:
* How would you comment on the pattern of online purchases across different generations.

iii. In your chart, further divide each distribution into subgroups of customers based on the **number of kids at home** (not including teens).
How would you comment on the pattern of online purchases across the different subgroups of customers?




```python
df_online_all = df_natadataset_copy.copy()
df_online_all['Generation'] = df_online_all['Year_Birth'].apply(lambda x:
    'Gen X' if 1965 <= x <= 1980 else
    'Millennials' if 1981 <= x <= 1996 else
    None)

df_online_all = df_online_all[df_online_all['Generation'].notna()]

fig1 = px.box(
    df_online_all,
    x='Generation',
    y='NumWebPurchases',
    points='outliers',
    color='Generation',
    title='Variability of Online Purchases by Generation')

fig1.update_layout(
    yaxis_title="Number of Online Purchases",
    xaxis_title="Generation",
    plot_bgcolor='white',
    showlegend=False)

fig1.show()
```

#**Part ii of C**
#### *Question 1:* How would you comment on the pattern of online purchases across different generations?

Gen X spends more overall than Millennials. Maybe, Nata's demographic is the older generation and something does not appeal as much to Millennials as much as it does for those born between 1965-1980.

```python
df_online_no_teens = df_natadataset_copy.copy()
df_online_no_teens['Generation'] = df_online_no_teens['Year_Birth'].apply(lambda x:
    'Gen X' if 1965 <= x <= 1980 else
    'Millennials' if 1981 <= x <= 1996 else
    None)

df_online_no_teens = df_online_no_teens[df_online_no_teens['Generation'].notna()]

fig2 = px.box(
    df_online_no_teens,
    x='Generation',
    y='NumWebPurchases',
    color='Kidhome',
    points='outliers',
    category_orders={'Kidhome': sorted(df_online_no_teens['Kidhome'].unique())},
    title='Variability of Online Purchases by Generation and Number of Kids')

fig2.update_layout(
    yaxis_title="Number of Online Purchases",
    xaxis_title="Generation",
    legend_title="Number of Kids at Home",
    plot_bgcolor='white')

fig2.show()
```
